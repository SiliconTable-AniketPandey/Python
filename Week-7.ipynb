```python
results = [ ] 
for i in range(8): 
    L = input().split(',') 
    winner = L[0]      
    losers = L[1: ]   
  
    results.append((winner, len(losers))) 
 
table = [ ] 

while results != [ ]: 
    max_team = results[0] 
    for i in range(len(results)): 
        team = results[i] 
        if team[1] > max_team[1]: 
            max_team = team 
        elif team[1] == max_team[1] and team[0] < max_team[0]: 
            max_team = team 
    results.remove(max_team)
    table.append(max_team) 
 
for team in table: 
    print(f'{team[0]}:{team[1]}') 
```

```python
def merge(D1, D2, priority): 
    if priority=='first': 
        for i in D2: 
            if i not in D1: 
                D1[i]=D2[i] 
        return D1 
    if priority=='second': 
        for i in D1: 
             if i not in D2: 
                D2[i]=D1[i]
    return D2 
```

```python
def minor_matrix(M, i, j):
    l=[] 
    for a in range(len(M)): 
        t=[] 
        if a==i: 
            continue  
        else:  
            for b in range(len(M[0])): 
                if b==j: 
                    continue 
                else: 
                    t.append(M[a][b]) 
        l.append(t) 
    return l 
```
