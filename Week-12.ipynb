```text

 Write a function named primes_galore that acepts a list of non-negative integers as arguments and returns the 
 numbers of primes that are located at prime indices in L, Zero-indexing is used throught this problem.

 For example:
 L = [1, 3, 11, 17, 23, 6, 8, 10]

 The prime indices in the list are 2,3,5,...7. of these, there are prime numbers at the indices 2 and 5. Therefore the
 Function should return the avlue 2 in this case

```

````python
def is_prime(num):
    if num <= 1:  # In question it is said that the number is not negative
        return False
    if num <= 3:  # prime numbers are those greater than 1 and only divisible by 1 and themselves.
        return True                   
    
    if num % 2 == 0 or num % 3 == 0: # Exclude multiples of 3 and 2 as they cannot be prime anyways
        return False
    
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i+2) == 0:
            return False
        i += 6
    return True

def prime_galore(l):
    prime_indices = []
    for i in range(len(l)):
        if is_prime(i):
            prime_indices.append(i)
            
    cout_of_primes = 0
    for indices in prime_indices:
        if indices < len(l) and  is_prime(l[indices]):
            cout_of_primes +=1
            
    return cout_of_primes

L = [1, 3, 11, 17, 23, 6, 8, 10]
result = prime_galore(L)
print("Number of primes at prime indices:", result)
        
````
