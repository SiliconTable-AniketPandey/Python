#### <u> Type conversion and type casting </u>


```python
a = 552.545
print("a = ", a, "The type of a is ", type(a))
print(
    "The value of a when converted to integer is ",
    int(a),
    "and now it is",
    type(int(a)),
)

b1 = bool(10)
b2 = bool(
    0
)  # Boolean value of 0 is always false and for any other value float or int it is always true.
b3 = bool(-30.52)


print("b1 = ", b1, "and type of b1 is ", type(b1))
print("b2 = ", b2, "and type of b2 is ", type(b2))
print("b3 = ", b3, "and type of b3 is ", type(b3))
print("\n\n")


b1 = bool("10")
b2 = bool("0")  # Boolean value of strings are always true except for empty strings.
b3 = bool("-30.52")
b4 = bool("Aniket Pandey")
b5 = bool("")  # Boolean value of an empty string is always false.

print("b1 = ", b1, "and type of b1 is ", type(b1))
print("b2 = ", b2, "and type of b2 is ", type(b2))
print("b3 = ", b3, "and type of b3 is ", type(b3))
print("b4 = ", b4, "and type of b4 is ", type(b4))
print("b5 = ", b5, "and type of b5 is ", type(b5))
```

##### Accept a sequence of five single digit numbers separated by commas as input. Print the product of all five numbers.

##### Input 1,2,3,4,5
##### Output 120

```python
numbers = input()
numberlist = []
numberlist = numbers.split(",")
print(numberlist)
result = 1
numbers = [int(i) for i in numberlist]

for i in numbers:
    result = result * i
print(result, end="")
```
