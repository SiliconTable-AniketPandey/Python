```python
def twin_primes(p,q):
    if p <0 or q <0:
        return False
    
    if abs(p -q )!=2:
        return False
    
    if p > q:
        for i in range(2 , int(p/2)):
            if p % i ==0 or q %2 ==0:
                return False
        return True
    
    else:
        for i in range(2 , int(q/2)):
            if q % i ==0 or p % i ==0:
                return False
        return True

twin_primes(13, 11)
    
```

```python
def long_tail(L):
    new_list=[]
    L=[str(numbers) for numbers in L ]
    t_counter = 0
    for i in L:
        s=i.split(".")    # type(s) = List ["12","1234"]
        if len(s[0]) < len(s[1]):
            t_counter +=1
    return t_counter

long_tail([12.1234, 155.231, 1981.1843213, 233.3333, 1001.1001, 100.0])
```
